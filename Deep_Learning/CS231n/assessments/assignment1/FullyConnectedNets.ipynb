## Inline Question 1:
Did you notice anything about the comparative difficulty of training the three-layer network vs. training the five-layer network? In particular, based on your experience, which network seemed more sensitive to the initialization scale? Why do you think that is the case?

## Answer:

Yes, there is a very noticeable difference. The **five-layer network is significantly more difficult to train and much more sensitive to the initialization scale** than the three-layer network.

### Why this is the case: The Problem of Vanishing & Exploding Gradients

This sensitivity is a classic problem in training deep neural networks, directly related to how gradients flow during backpropagation.

Backpropagation relies on the chain rule, meaning that to update the weights in the *first* layer, you have to multiply the gradient signal backward through *all* the layers that come after it (layers 5, 4, 3, and 2).

This creates a "compounding" effect based on the magnitude of the weights at each layer:

1.  **If `weight_scale` is too small:**
    * The weights ($W$) will likely be, on average, less than 1.
    * During backpropagation, the gradient is repeatedly multiplied by these small weights (e.g., $grad \times W_5 \times W_4 \times W_3 \times W_2$).
    * This causes the gradient signal to shrink exponentially as it travels backward.
    * By the time the gradient reaches the early layers (like layer 1), it's practically zero.
    * This is called the **vanishing gradient problem**. The early layers, which are crucial for learning basic features, stop learning entirely, and the network fails to converge.

2.  **If `weight_scale` is too large:**
    * The weights will be, on average, greater than 1.
    * During backpropagation, the gradient is repeatedly multiplied by these large numbers.
    * This causes the gradient signal to grow exponentially.
    * This is called the **exploding gradient problem**. The gradients become massive (or even `NaN` - Not a Number), causing the weight updates to be huge and unstable. The loss will skyrocket, and the model completely diverges.

The **three-layer network** is less sensitive because it's "shallower." It only has a couple of multiplications in its chain rule, so this compounding effect is much weaker. It can tolerate a wider, less-perfect range of initialization scales.

In contrast, the **five-layer network's** greater depth *amplifies* any small error in the initialization scale, making it fail dramatically if the scale isn't "just right."

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
**1. Why do AdaGrad updates become very small?**

This is entirely due to AdaGrad's `cache` update rule: `cache += dw**2`.

* **`cache` is an accumulator:** The `cache` variable **accumulates** the squared gradients (`dw**2`) from *all* past iterations since the beginning of training.
* **`cache` is monotonically increasing:** Because `dw**2` is always non-negative, the value of `cache` will only get larger and larger throughout training—it *never decreases*.
* **`cache` is in the denominator:** In the weight update rule `w += ...`, the `np.sqrt(cache)` term is in the denominator.
* **Result:** As training progresses, this denominator `np.sqrt(cache)` continuously grows larger. This causes the entire update step (the effective learning rate) to perpetually decay, eventually becoming **extremely small and approaching zero**.

This is why John's network stopped learning: the update steps virtually vanished, and the weights almost stopped changing.

**2. Would Adam have the same issue?**

**No**, Adam (and RMSProp) cleverly solves this exact problem.

* **Adam uses an "Exponentially Decaying Moving Average":** Adam does *not* accumulate *all* past squared gradients. Its cache for squared gradients (typically called `v`) follows this rule:
    `v = beta2 * v + (1 - beta2) * (dw**2)`
* **Focus on recent gradients:** The `beta2` term (usually ~0.999) is a "decay rate." This formula means `v` is a **moving average** that gives more weight to *recent* gradients and gradually "forgets" gradients from the distant past.
* **The denominator does not grow indefinitely:** Because `v` is just an *estimate* of the recent gradient magnitude, not an ever-growing sum, the denominator in the Adam update rule does not grow uncontrollably like AdaGrad's.

**In summary:**
AdaGrad's learning rate aggressively decays to zero because it accumulates *all* history. Adam's learning rate adapts based on a *recent, moving average* of gradient history, allowing it to continue learning effectively without its update steps vanishing prematurely.
