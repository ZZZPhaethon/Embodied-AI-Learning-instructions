# Fully-Connected Neural Nets
In this exercise we will implement fully-connected networks using a modular approach. For each layer we will implement a `forward` and a `backward` function. The `forward` function will receive inputs, weights, and other parameters and will return both an output and a `cache` object storing data needed for the backward pass, like this:

```python
def layer_forward(x, w):
  """ Receive inputs x and weights w """
  # Do some computations ...
  z = # ... some intermediate value
  # Do some more computations ...
  out = # the output
   
  cache = (x, w, z, out) # Values we need to compute gradients
   
  return out, cache
```

The backward pass will receive upstream derivatives and the `cache` object, and will return gradients with respect to the inputs and weights, like this:

```python
def layer_backward(dout, cache):
  """
  Receive dout (derivative of loss with respect to outputs) and cache,
  and compute derivative with respect to inputs.
  """
  # Unpack cache values
  x, w, z, out = cache
  
  # Use values in cache to compute derivatives
  dx = # Derivative of loss with respect to x
  dw = # Derivative of loss with respect to w
  
  return dx, dw
```

After implementing a bunch of layers this way, we will be able to easily combine them to build classifiers with different architectures.
  

````python
def layer_forward(x, w):
    """ Receive inputs x and weights w """
    # Step 1: compute intermediate result
    z = np.dot(x, w)

    # Step 2: maybe apply activation
    out = np.maximum(0, z)  # example if using ReLU

    # Step 3: save necessary values for backward pass
    cache = (x, w, z, out)

    return out, cache
````

````python
def layer_backward(dout, cache):
    """ Compute gradients using cached values """
    # Unpack cache
    x, w, z, out = cache

    # Derivative computations
    dz = dout * (z > 0)    # if ReLU
    dw = x.T.dot(dz)
    dx = dz.dot(w.T)

    return dx, dw
````

## Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

$\color{blue}{\textit Your Answer:}$ *$\color{blue}{\textit{Your Answer:}}$
The following two activation functions have the problem of zero (or near-zero) gradients:

1.  **Sigmoid**
    * **Problem Description:** The Sigmoid function **saturates** when the absolute value of its input is very large.
    * **Input that causes this problem:** When the input value is **very large ($\to \infty$)** or **very small ($\to -\infty$)**. In the 1D case, when $x$ is a large positive or large negative number, the Sigmoid function's curve becomes very flat, and its derivative (gradient) approaches 0. This leads to the **Vanishing Gradient Problem**, where gradients become minuscule after backpropagating through many layers, causing the network to learn very slowly or stop.

2.  **ReLU (Rectified Linear Unit)**
    * **Problem Description:** ReLU has a gradient of exactly zero for half of its domain, which can lead to the "**Dying ReLU**" problem.
    * **Input that causes this problem:** When the input value is **negative ($x < 0$)**. For all negative inputs, the output of ReLU is 0, and its local gradient is also 0. If a neuron (due to, for example, a large learning rate or parameter shift) consistently receives only negative inputs during training, it will never activate, and its weights will never be updated because the gradient is always 0.

---

**Leaky ReLU** was designed to address the "Dying ReLU" problem. For negative inputs ($x < 0$), it returns a small, non-zero value (e.g., $\alpha x$, where $\alpha=0.01$), so its gradient is a small positive constant $\alpha$ instead of 0. This ensures that a gradient always exists, even in the negative region, preventing the neuron from "dying".*

````python
#################################################################################
# TODO: Tune hyperparameters using the validation set. Store your best trained  #
# model in best_model.                                                          #
#                                                                               #
# To help debug your network, it may help to use visualizations similar to the  #
# ones we used above; these visualizations will have significant qualitative    #
# differences from the ones we saw above for the poorly tuned network.          #
#                                                                               #
# Tweaking hyperparameters by hand can be fun, but you might find it useful to  #
# write code to sweep through possible combinations of hyperparameters          #
# automatically like we did on the previous exercises.                          #
#################################################################################

best_model = None
best_val_acc = -1.0  # 初始化最佳验证准确率
results = {}         # 存储 (超参数) -> (验证准确率) 映射

# --- 定义超参数搜索范围 ---
# 我们将使用随机搜索。
num_trials = 20  # 你可以增加这个数字以获得更好的结果，但这会花费更长时间

# 学习率 (lr) 和 正则化强度 (reg) 在对数尺度上采样更有效
# 隐藏层大小 (hs) 在线性尺度上采样

lr_range = [-3.5, -2.5]  # 学习率范围: 10^-3.5 (~3e-4) 到 10^-2.5 (~3e-3)
reg_range = [-2, 0]      # 正则化强度范围: 10^-2 (0.01) 到 10^0 (1.0)
hs_range = [50, 200]     # 隐藏层大小范围: 50 到 200

print(f"Starting random search for {num_trials} trials...")

for i in range(num_trials):
    # 1. 随机采样超参数
    lr = 10**np.random.uniform(lr_range[0], lr_range[1])
    reg = 10**np.random.uniform(reg_range[0], reg_range[1])
    hidden_size = np.random.randint(hs_range[0], hs_range[1])

    # 2. 创建模型
    model = TwoLayerNet(input_size, hidden_size, num_classes,
                        weight_scale=1e-3, reg=reg)

    # 3. 创建 Solver 并训练
    solver = Solver(model, data,
                    update_rule='sgd',
                    optim_config={
                        'learning_rate': lr,
                    },
                    lr_decay=0.95,
                    num_epochs=10,  # 使用较少的 epoch 进行快速搜索
                    batch_size=200,
                    verbose=False) # 在搜索时关闭详细输出

    solver.train()

    # 4. 获取最佳验证准确率
    val_acc = solver.best_val_acc

    # 5. 存储结果
    results[(lr, reg, hidden_size)] = val_acc

    print(f"Trial {i+1}/{num_trials}: lr={lr:.1e}, reg={reg:.2f}, hs={hidden_size} ==> val_acc: {val_acc:.4f}")

    # 6. 检查这是否是迄今为止最好的模型
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_model = model  # 关键：存储训练好的模型对象

# --- 搜索结束 ---

print("\n-------------------------------------------------")
print("Hyperparameter search finished.")
print(f"Best validation accuracy achieved: {best_val_acc:.4f}")

# 找到并打印最佳超参数组合
best_params_combo = max(results, key=results.get)
print(f"Best hyperparameters (lr, reg, hs): {best_params_combo}")
print("-------------------------------------------------")


################################################################################
#                              END OF YOUR CODE                                #
################################################################################
````

Now that you have trained a Neural Network classifier, you may find that your testing accuracy is much lower than the training accuracy. In what ways can we decrease this gap? Select all that apply.

1. Train on a larger dataset.
2. Add more hidden units.

$\color{blue}{\textit Your Answer:}$ *1, 3*

$\color{blue}{\textit Your Explanation:}$
A large gap between training and testing accuracy (where training accuracy is much higher) is the classic sign of overfitting. The model has learned to "memorize" the training data, including its noise, instead of learning the underlying general patterns. Our goal is to improve the model's generalization.

Train on a larger dataset (Correct): This is one of the most effective ways to combat overfitting. By providing more diverse examples, you make it much harder for the model to simply memorize the data. It is forced to learn the true, generalizable patterns that are common across all the data, which improves its performance on the unseen test set.

Increase the regularization strength (Correct): Regularization (like L2 regularization) adds a penalty term to the loss function based on the magnitude of the model's weights. Increasing this strength forces the model to use smaller weights, effectively "simplifying" the model and preventing it from fitting the training data's noise too closely. This improves generalization and reduces the gap.

Add more hidden units (Incorrect): Adding more hidden units increases the model's capacity (complexity). A more complex model has more power to memorize the training data, which would likely increase overfitting and make the gap between training and testing accuracy even larger.
